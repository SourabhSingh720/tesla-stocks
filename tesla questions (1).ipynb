15. Write a Python program to store the values of Date(s) and Volume attributes in a dataframe where the Volume of shares traded is between 1000000 and 1500000. Display the dataframe as follows.

```python
df_volume=df.loc[(df.Volume>1000000) & (df.Volume<1500000)][['Date','Volume']].reset_index(drop=True)
df_volume
```

16. Write a Python program to draw a bar plot of Date versus Volume values using the dataframe created in (15) as depicted below.

```python
plt.figure(figsize=(10,5))
df_volume['Date']=pd.to_datetime(df_volume['Date']).dt.date
sns.barplot(data=df_volume,x='Date',y='Volume')
plt.xticks(rotation=90)
```
